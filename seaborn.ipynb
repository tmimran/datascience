```python
# Seaborn - Data Visualization Library

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# built-in datasets of seaborn
sns.get_dataset_names()
```

```python
dataset = sns.load_dataset('iris')
dataset
```

```python
# lineplot() used to generate a line graph
sns.lineplot(x='sepal_length', y='sepal_width',data=dataset)
plt.title('Line Graph using Seaborn')
plt.show()
```

```python
# adding style to seaborn
sns.set_style('whitegrid',{'grid.color': 'k','grid.linestyle':':'}) #linestyle -- : -
sns.lineplot(x='sepal_length', y='sepal_width',data=dataset)
plt.title('Line Graph using Seaborn')
plt.xticks(fontsize=10) # xticks - perkataan nilai x
plt.yticks(fontsize=10) # yticks - perkataan nilai y
plt.show()
```

```python
# change the figure size
plt.figure(figsize=(10,5)) 
sns.set_style('whitegrid',{'grid.color': 'k','grid.linestyle':':'}) #linestyle -- : -
sns.lineplot(x='sepal_length', y='sepal_width',data=dataset)
plt.title('Line Graph using Seaborn')
plt.xticks(fontsize=10) # xticks - perkataan nilai x
plt.yticks(fontsize=10) # yticks - perkataan nilai y
plt.show()
```

```python
pokemon = pd.read_csv('https://raw.githubusercontent.com/yash240990/Python/master/pokemon.csv')
pokemon
```

```python
# scatterplot()
sns.scatterplot(x='Attack',y='Defense',data=pokemon)
plt.title('Scatter Plot using Seaborn')
plt.show()
```

```python
pokemon.Generation.value_counts().sort_index()
```

```python
# hue = category
# htmlcolorcodes.com --> provide good color for blending
sns.scatterplot(x='Attack', y='Defense', hue='Generation', data=pokemon)
plt.title('Scatter Plot using Seaborn')
plt.show()
```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
sns.scatterplot(x='Attack', y='Defense', hue='Generation', palette=colors,data=pokemon)
plt.title('Scatter Plot using Seaborn')
plt.show()
```

```python
# lmplot()
# used to generate a scatter plot using a regression line
# Help machine learning berkenaan precision
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
sns.lmplot(x='Attack', y='Defense', hue='Generation', palette=colors,data=pokemon)
plt.title('Scatter Plot using Seaborn')
plt.show()
```

```python
# countplot()
# used to generate a bar graph

pokemon['Type 1'].value_counts()
```

```python
# countplot()
# used to generate a bar graph
plt.figure(figsize=(15,5))
sns.countplot(x='Type 1',data=pokemon)
plt.title('Count Plot using Seaborn')
plt.show()
```

```python
#import warnings
#warnings.filterwarnings('ignore')
'''Just in case warningnya bukan error'''

pkmn_type_colors = ['#3ae40c',  # Grass
                    '#e52106',  # Fire
                    '#0AE3D6',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#ECF709',  # Dragon
                    '#7A786C',  # Dark
                    '#D3D1C0',  # Steel
                    '#91E8EE',  # Flying
                   ]

plt.figure(figsize=(10,5)) #dont right huge figure size
plt.xticks(rotation=-45) #tak nak naming 'x' overlap
sns.countplot(x='Type 1', hue='Type 1',palette=pkmn_type_colors,data=pokemon)
plt.title('Count Plot using Seaborn')
plt.show()
```

```python
tips = sns.load_dataset('tips') # dataset daripada seaborn sendiri
tips
```

```python
# When relationship type dont put title, make more confusing
# What time my waiter might get tips more during dinner or lunch and what day?
sns.relplot(x='total_bill',y='tip',col='time',hue='day',data=tips)
plt.show()
```

```python
# Who provides tips more, male or female?
sns.relplot(x='total_bill',y='tip',col='sex',data=tips)
plt.show()
```

```python
# Who provides tips more, male or female? is he/she smoker?
sns.relplot(x='total_bill',y='tip',col='sex',hue='smoker',data=tips)
plt.show()
```
